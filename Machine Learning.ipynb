# SIMPLE LINEAR REGRESSION

```python
#STEP 1: IMPORT LIBRARY
import pandas as pd
```

```python
#STEP 2: READ DATASET
df=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv');

```

```python
df.head();
```

```python
df.info();
```

```python
df.describe();
```

```python
#STEP 3: DEFINE Y AND X

df.columns;

y = df['Salary'];
X = df[['Experience Years']];
```

```python
#STEP 4: TRAIN-TEST-SPLIT
from sklearn.model_selection import train_test_split;

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529); #random_state=2529 is used so that sample space is not changed.We can use any number after =.
X.shape, X_train.shape, X_test.shape
```

```python
#STEP 5: SELECT MODEL
from sklearn.linear_model import LinearRegression
model = LinearRegression();
```

```python
#STEP 6: TRAIN MODEL
model.fit(X_train,y_train)
model.intercept_
model.coef_
#with intercept_ and coef_, we get the equation y=b0+b1X or y=c+mX
```

```python
#STEP 7: PREDICTION
y_pred = model.predict(X_test)
model.predict([[9.2]])
```

```python
#STEP 8: EVALUATION
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)

model.predict([[9.2]])

import seaborn as sns
sns.regplot(x='Experience Years', y='Salary', data=df) #gives the scatter plot with regression
```

# BINARY CLASSIFICATION WITH LOGISTIC REGRESSION

```python
import pandas as pd
```

```python
df=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
y=df['diabetes']
X=df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
import seaborn as sns
sns.regplot(x=df.glucose,y=df.diabetes,data=df)
```

```python
#Here, we can see that when we use linear regression, it gives us a meaningless data.
#i.e, diabetes can hold only values 0 and 1. But for a patient with glucose 25, it gives diabetes -0.2.

#So, we'll be using Logistic Regression where the equation is:
# P(y)=1/(1+exp^-(b0+b1X))

# i.e, we used sigmoid/logistic transformation and it gives O/P as a probab.
```

```python
import seaborn as sns
sns.regplot(x=df.glucose,y=df.diabetes,data=df,logistic=True)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=500) #we need to increase the no. of iterations to remove any warnings
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

# NAIVE BAYES CLASSIFIERS

```python
#The major difference b/w NB & Bayes is that, here we do not calculate the P(x) in the denominator.
#We are concerned about the output only and not the probab.

import pandas as pd
```

```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Bank%20Churn%20Modelling.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#We need to do some pre processing.
#CustomerId and Surname are irrelevant info for us, so we'll drop it.
#Geography and Gender are in the form of text, so we need to convert it into numbers format
```

```python
df.columns
```

```python
df = df.drop(['CustomerId', 'Surname'],axis=1)
```

```python
df.Geography.value_counts()
```

```python
df.Gender.value_counts()
```

```python
pd.get_dummies(df)
#if the geography is France,Germany or Spain, then it shows true(1) otherwise false(0)
#similarly for gender also
```

```python
df = pd.get_dummies(df)
```

```python
df.columns
```

```python
y = df['Churn']
X = df[['CreditScore', 'Age', 'Tenure', 'Balance', 'Num Of Products',
       'Has Credit Card', 'Is Active Member', 'Estimated Salary',
       'Geography_France', 'Geography_Germany', 'Geography_Spain',
       'Gender_Female', 'Gender_Male']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

# DECISION TREE

```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
y = df['diabetes']
X = df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
dtc = DecisionTreeClassifier(max_depth=3)  #defines the no. of splitting or branches in the tree
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#plot_tree(dtc);
plot_tree(dtc,feature_names=X.columns);
```
