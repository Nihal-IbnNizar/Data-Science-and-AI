## SIMPLE LINEAR REGRESSION

```python
#STEP 1: IMPORT LIBRARY
import pandas as pd
```

```python
#STEP 2: READ DATASET
df=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv');

```

```python
df.head();
```

```python
df.info();
```

```python
df.describe();
```

```python
#STEP 3: DEFINE Y AND X

df.columns;

y = df['Salary'];
X = df[['Experience Years']];
```

```python
#STEP 4: TRAIN-TEST-SPLIT
from sklearn.model_selection import train_test_split;

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529); #random_state=2529 is used so that sample space is not changed.We can use any number after =.
X.shape, X_train.shape, X_test.shape
```

```python
#STEP 5: SELECT MODEL
from sklearn.linear_model import LinearRegression
model = LinearRegression();
```

```python
#STEP 6: TRAIN MODEL
model.fit(X_train,y_train)
model.intercept_
model.coef_
#with intercept_ and coef_, we get the equation y=b0+b1X or y=c+mX
```

```python
#STEP 7: PREDICTION
y_pred = model.predict(X_test)
model.predict([[9.2]])
```

```python
#STEP 8: EVALUATION
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)

model.predict([[9.2]])

import seaborn as sns
sns.regplot(x='Experience Years', y='Salary', data=df) #gives the scatter plot with regression
```

## BINARY CLASSIFICATION WITH LOGISTIC REGRESSION

```python
import pandas as pd
```

```python
df=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
y=df['diabetes']
X=df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
import seaborn as sns
sns.regplot(x=df.glucose,y=df.diabetes,data=df)
```

```python
#Here, we can see that when we use linear regression, it gives us a meaningless data.
#i.e, diabetes can hold only values 0 and 1. But for a patient with glucose 25, it gives diabetes -0.2.

#So, we'll be using Logistic Regression where the equation is:
# P(y)=1/(1+exp^-(b0+b1X))

# i.e, we used sigmoid/logistic transformation and it gives O/P as a probab.
```

```python
import seaborn as sns
sns.regplot(x=df.glucose,y=df.diabetes,data=df,logistic=True)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=500) #we need to increase the no. of iterations to remove any warnings
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

## NON PARAMETRIC [kNearestNeighbor-Regression]

```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Admission%20Chance.csv')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
y = df['Chance of Admit '];
X = df[['GRE Score', 'TOEFL Score', 'University Rating', ' SOP',
       'LOR ', 'CGPA', 'Research']];
```

```python
from sklearn.model_selection import train_test_split;

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529);
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors=10);
```

```python
knr.fit(X_train,y_train)
```

```python
y_pred = knr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```
